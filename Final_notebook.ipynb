### Revenue Per Upload/Subscriber by Category

## Q5. Do higher urban populations correlate with view and subscriber count?

Relevant Columns
- Urban_population: Represents the urban population of the country associated with the Youtuber.
- Subscribers: Number of subscribers for the Youtuber.
- Video views: Total video views.

```python
data = youtube[['Urban_population', 'subscribers', 'video views']].dropna()
```

```python
data.info()
```

### **1. Scatter Plot: Urban Population vs. Subscribers**

This plot shows the relationship between the size of the urban population and the number of subscribers. Points suggest whether urbanization levels might influence audience size.

```python
# Scatter Plot: Urban Population vs. Subscribers

import matplotlib.pyplot as plt
import seaborn as sns

# Scatter Plot (Urban Population vs Subscribers)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Urban_population', y='subscribers', alpha=0.7, color='blue')
plt.title("Urban Population vs Subscribers")
plt.xlabel("Urban Population")
plt.ylabel("Subscribers")
plt.grid(True)
plt.show()
```

- **Findings**:  
  The plot displays a spread of points, but no clear upward or downward trend is apparent. This suggests that an increase in the urban population does not consistently result in a higher number of subscribers.
- **Interpretation**:  
  The absence of a strong pattern implies that the size of the urban population may not be a significant factor influencing subscriber count.

### **2. Scatter Plot: Urban Population vs. Video Views**

This visualization examines how urban population size correlates with total video views, highlighting potential trends

```python
# Scatter Plot: Urban Population vs. Video Views
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Urban_population', y='video views', alpha=0.7, color='green')
plt.title("Urban Population vs Video Views")
plt.xlabel("Urban Population")
plt.ylabel("Video Views")
plt.grid(True)
plt.show()
```

- **Findings**:  
  Similar to the first scatter plot, the data points are widely scattered with no discernible trend or clustering. There is no consistent increase in video views with larger urban populations.
- **Interpretation**:  
  This indicates that video views are also not strongly correlated with the size of the urban population.

### **3. Correlation Heatmap**

Displays correlation coefficients among urban population, subscribers, and video views. Positive values suggest a direct relationship, while negative ones suggest an inverse relationship.

```python
# Correlation Heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

- **Findings**:  
  The heatmap shows the correlation coefficients:
  - Urban Population vs. Subscribers: Near zero or very low positive/negative values.
  - Urban Population vs. Video Views: Similarly weak correlation values.
- **Interpretation**:  
  These values confirm that the relationships between urban population, subscriber count, and video views are either negligible or non-existent.
  

### **4. 3D Scatter Plot: Urban Population, Subscribers, and Video Views**

Allows dynamic exploration of how urban population, subscribers, and video views interact in three dimensions.

```python
import plotly.express as px

# 3D Scatter Plot
fig = px.scatter_3d(data, x='Urban_population', y='subscribers', z='video views',
                    color='Urban_population',
                    title="3D Scatter Plot: Urban Population, Subscribers, and Video Views",
                    labels={"Urban_population": "Urban Population", "subscribers": "Subscribers", "video views": "Video Views"})
fig.show()
```

- **Findings**:  
  The interactive visualization highlights no significant clustering or linear patterns connecting urban population with both video views and subscribers simultaneously.
- **Interpretation**:  
  This corroborates earlier findings, as urban population size appears to have little influence on both metrics.



### **Conclusion**
From the analysis:
- **Higher urban populations do not strongly correlate with subscriber counts or video views**.
- The scatter plots, heatmap, and 3D visualization all indicate that other factors are likely driving differences in subscriber counts and video views, rather than urban population size.


## Regression

```python
youtube.loc[:,'viewtosub'] = youtube['video views'] / youtube['subscribers']
```

```python
# Regression on unfiltered Data
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Add a constant (intercept) term to the independent variable
X = sm.add_constant(youtube['viewtosub'])  # Adds the intercept
y = youtube['highest_yearly_earnings']

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())

# Plot the data and regression line
plt.scatter(youtube['viewtosub'], youtube['highest_yearly_earnings'], label='Data points',alpha=0.5)
plt.plot(youtube['viewtosub'], model.predict(X), color='red', label='Regression line')
plt.title('Unfiltered ViewstoSubs vs Highest Yearly Earnings')
plt.xlabel('Views to Subscribers')
plt.ylabel('Highest Yearly Earnings (hundred million)')
plt.legend()
plt.show()
```
